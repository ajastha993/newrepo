Ploting a line using plot


```
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x=[1,2,3]
y=[4,5,1]
plt.plot(x,y)
plt.title("graph")
plt.show()
```

```
plt.bar(x,y,color='g')
plt.bar([1,3,9],[4,2,3])
plt.legend()
```

```
plt.hist([1,20,33,4,59,16,78,8,9],[0,10,20,30,40,50,60,70,80],histtype='bar')
```

```
plt.scatter(x,y)
```

```
plt.stackplot(x,y)
```

```
plt.pie(x,y)
```
